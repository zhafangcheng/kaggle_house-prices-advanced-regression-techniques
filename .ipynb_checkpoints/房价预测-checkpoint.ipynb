```python
# 导入基本数据包
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore") # 忽略警告
```

```python
# 读取数据（包括训练集数据与测试集数据）
train_df = pd.read_csv('/Users/fangcheng/数据挖掘/house-prices-advanced-regression-techniques/train.csv',index_col=0)
test_df = pd.read_csv('/Users/fangcheng/数据挖掘/house-prices-advanced-regression-techniques/test.csv',index_col=0)
```

```python
# 查看数据
train_df.head()
```

```python
# 合并训练集与测试集，先抽取出目标变量
# log1p的目的：让目标变量数据符合正态分布
# 如果使用了正态分布数据化后，最终的结果需要把正态分布的数据还原
prices = pd.DataFrame({"price":train_df["SalePrice"],"log(price + 1)":np.log1p(train_df["SalePrice"])})
prices.hist()
```

```python
# 抽取变量数据
y_train = np.log1p(train_df.pop("SalePrice"))
```

```python
# 训练集与测试集数据合并
all_df = pd.concat((train_df,test_df), axis=0)
```

```python
all_df.shape
```

```python
y_train.head()
```

```python
# 变量转化//特征工程
# MSSubClass只是一个MSSubClass
all_df["MSSubClass"].dtypes
```

```python
all_df["MSSubClass"] = all_df["MSSubClass"].astype(str)
```

```python
# 统计每一个类别的个数
all_df["MSSubClass"].value_counts()
```

```python
# 把category的变量转化成unmerical表达形式，变成one-hot
pd.get_dummies(all_df["MSSubClass"],
prefix = 'MSSubClass').head()
```

```python
# 上一个步骤把MSSubClass分成12个column，每一个代表一个category。是就是1，不是就是0
```

```python
# 把所有的category数据，都给one-hot
all_dummy_df = pd.get_dummies(all_df)
all_dummy_df.head()
```

```python
# 处理数据缺失情况
all_dummy_df.isnull().sum().sort_values(ascending=False).head()
```

```python
# 可以使用多种方法来填充数据的缺失，此案例先采用均值填充(0、最大值-最小值、中位数等)
mean_cols = all_dummy_df.mean()
mean_cols.head(10)
```

```python
all_dummy_df = all_dummy_df.fillna(mean_cols)
```

```python
all_dummy_df.isnull().sum().sum()
```
